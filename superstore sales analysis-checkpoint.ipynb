```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\admin\OneDrive\Desktop\Project CSV\train.csv")
df.head()
```

```python
df.info()
```

```python
# null values 
null_count = df['Postal Code'].isnull().sum()
null_count
```

```python
#fill NAN with 0
df['Postal Code'].fillna(0,inplace=True)
```

```python
df.info()
```

```python
#change folat to integer
df['Postal Code'] = df['Postal Code'].astype(int)
```

```python
#check for duplicates
df.duplicated(keep=False).sum()
```

```python
#let's check for the types of customers
type_of_customers = df['Segment'].unique()
type_of_customers
```

```python
#no.of customers
no_of_customers = df['Segment'].value_counts().reset_index()
no_of_customers
```

```python
no_of_customers = no_of_customers.rename(columns={'index':'Customer type','Segment':'Total customers'})
no_of_customers
```

```python
#pie chart
plt.pie(no_of_customers['Total customers'],labels=no_of_customers['Customer type'], autopct='%1.1f%%')
plt.title('Distribution of customers')
plt.show()
```

```python
sales_category = df.groupby('Segment')['Sales'].sum().reset_index()
sales_category = sales_category.rename(columns={'Segment':'Customer type','Sales':'Total sales'})

sales_category
```

```python
#pie chart
plt.pie(sales_category['Total sales'],labels=sales_category['Customer type'],autopct='%1.1f%%')
plt.title('Sales per customer category')
plt.show()
```

```python
# customer loyalty
customer_order_freq = df.groupby(['Customer ID','Customer Name','Segment'])['Order ID'].count().reset_index()
customer_order_freq
```

```python
#rename order id column
customer_order_freq.rename(columns={'Order ID':'Total orders'},inplace=True)
```

```python
# identify repeated customers
repeat_customers = customer_order_freq[customer_order_freq['Total orders']>=1]
repeat_customers
```

```python
#sort repeat customers
sorted_repeat_customers = repeat_customers.sort_values(by='Total orders',ascending=False)
print(sorted_repeat_customers.head(10).reset_index(drop=True))
```

```python
# ranking customers according to sales
customer_sales = df.groupby(['Customer ID','Customer Name','Segment'])['Sales'].sum().reset_index()
customer_sales

#sort in descending order
top_spenders = customer_sales.sort_values(by='Sales',ascending=False)

print(top_spenders.head(10).reset_index(drop=True))
```

```python
# convert order date to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'],dayfirst=True)
```

```python
# grouping by year and summing the sales per year
yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum()
yearly_sales = yearly_sales.reset_index()
yearly_sales = yearly_sales.rename(columns={'Order Date':'Year','Sales':'Total_sales'})
yearly_sales
```

```python
# line graph
plt.plot(yearly_sales['Year'],yearly_sales['Total_sales'],marker='o',linestyle='--')
plt.title('Yearly sales')
plt.xlabel('Year')
plt.ylabel('Total sales')
plt.xticks(rotation=65)
plt.show()
```

```python
# convert to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'],dayfirst=True)
```

```python
#filtering by year
year_sales = df[df['Order Date'].dt.year==2018]
```

```python
# calculate quarterly sales
quarterly_sales = year_sales.resample('Q',on='Order Date')['Sales'].sum()
quarterly_sales = quarterly_sales.reset_index()
quarterly_sales = quarterly_sales.rename(columns={'Order Date':'Quarter','Sales':'Total sales'})
print('This are the quarterly sales for 2018')
print('quarterly_sales')
```

```python
# line graph
plt.plot(quarterly_sales['Quarter'],quarterly_sales['Total sales'],marker='o',linestyle='--')
plt.title('Quarterly sales')
plt.xlabel('Quarters')
plt.ylabel('Total sales')
plt.xticks(rotation=65)
plt.show()
```

```python
# monthly sales trend - 2018
df['Order Date'] = pd.to_datetime(df['Order Date'],dayfirst=True)
year_sales = df[df['Order Date'].dt.year==2018]
```

```python
monthly_sales = year_sales.resample('M',on='Order Date')['Sales'].sum()
monthly_sales = monthly_sales.reset_index()
monthly_sales = monthly_sales.rename(columns={'Order Date':'Month','Sales':'Total monthly sales'})
monthly_sales
```

```python
# line graph
plt.plot(monthly_sales['Month'],monthly_sales['Total monthly sales'],marker='o',linestyle='--')
plt.title('Monthly sales')
plt.xlabel('Months')
plt.ylabel('Total sales')
plt.xticks(rotation = 65)
plt.show()
```
